La librairie Pycaret permet de construire plus rapidement et efficacement des modèles de machine learning. Les fonctions créent, comparent, ajustent, et enregistrent les modèles de façon automatique

DATASET

```python
# Import des librairies et du dataset
import pandas as pd
from pycaret.regression import *
dataset= pd.read_csv('storage/best_cars.csv')
# Format du dataset (lignes, colonnes) et apercu
print(dataset.shape)
dataset.head()
```

Afin de réduire le temps d'apprentissage de nos modèles, je vais reduire le dataset original. 15% sera gardé, soit environ 8200 enregristrements. La cross-validation se fera avec k=5  
Je vais ensuite diviser le dataset réduit en 2 parties. 90% me servira à la modélisation (data)  
Data_unseen sera le reste sur lequel je testerai mes modèles. Ces observations seront 
considérées comme des nouvelles données pour les modèles

```python
dataset_reduced= dataset.sample(frac=0.15, random_state=786)

data = dataset_reduced.sample(frac=0.9, random_state=786)
data_unseen = dataset_reduced.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

# Format des 2 datasets (lignes, colonnes)
print('Data pour Modelisation: ' + str(data.shape))
print('Unseen Data pour les Predictions: ' + str(data_unseen.shape))
```

#  EXPERIENCE I: 

La 1ère expérimentation est faite sans aucun preprocessing.

##  *CONFIGURATION*

```python
# Configuration de la regression avec Pycaret
exp_reg101 = setup(data = data, target = 'price', session_id=123)
```

## *CONSTRUCTION ET COMPARAISON DES MODELES*

```python
# Comparaison des modeles avec cross-validation {k-fold=5}
compare_models(exclude= ['ransac', 'tr'],fold=5)
```

 Les métriques sur lesquelles je vais me focaliser sont le R2 (max), la RMSE et RMSLE (min)  
 Je vais donc porter mon attention ici sur 2 modèles en particulier  
 - catboost (meilleurs r2 et RMSE)
 - extra trees (meilleure RMSLE)

### 1.2.1 Catboost

```python
# Creation du modele
cat= create_model('catboost',fold= 5)
```

```python
# Fine tuning
tuned_cat = tune_model(cat, fold=5)
```

Le fine tuning n'a pas amelioré le modèle 

```python
# Prediction sur le test set
predict_model(cat)
```

### 1.2.2 Extra Tree Regressor

```python
# Creation du modele
extra_tree = create_model('et', fold=5)
```

```python
# Fine tuning
tuned_et= tune_model(extra_tree, fold=5)
```

Le fine tuning rend un modèle équivalent

```python
# Visualisation de l'importance des features
plot_model(tuned_et, plot='feature')
```

Il est intéressant de voir que les colonnes les plus importantes sont les mêmes présumées dans la partie 1 EDA

```python
# Prediction sur test-set
predict_model(extra_tree)
```

## *CHOIX DU MODELE*

Le modèle catboost a de meilleures performances sur le test set que le second modèle  
- R2 :	0.8322 	> 0.8203
- RMSE : 3494.0369 < 3616.7101 	 	
- RMSLE : 0.3996 < 0.4018

```python
# Modele finalise 
cat_final = finalize_model(cat)
```

##  *PREDICTION FINALE*

```python
# Predictions sur nouvelles observations
new_pred = predict_model(cat_final, data= data_unseen)
new_pred
```

```python
# Verification des metriques
from pycaret.utils import check_metric
print('R2 du modele :', check_metric(data_unseen.price, new_pred.Label, 'R2'))
print('RMSLE du modele :', check_metric(data_unseen.price, new_pred.Label, 'RMSLE'))
```

```python
# Sauvegarde du modele
save_model(cat_final, 'firt_cars_model_r84')
```

  Dans cette 1ère expérience, le modele choisit a été le Catboost regressor avec pour métriques sur de nouvelles données:

# EXPERIENCE II:

La 2nde expérience consistera à appliquer plusieurs modifications au dataset. En effet dans un cas de régression, plusieurs algorithmes tels que le KNN, Linear Regression ont besoin que le dataset remplisse certaines conditions afin de pouvoir mieux performer:    
 - normalize: les valeurs numériques sont centrées autour de 0 et ont des variances de même ordre de grandeur
 - transformation, transform_target: méthode qui consiste à donner aux variables, une distribution suivant la loi normale
 - combine_rare_levels : certaines variables categorielles peuvent avoir des fréquences tres faibles (ex: voitures à 12 cylindres, voitures électriques...) Il s'agit de les rassembler dans une nouvelle catégorie
 - ...

## *CONFIGURATION*

```python
# Configuration
exp_reg102 = setup(data = data, target = 'price', session_id=123, normalize = True,
                   transformation = True, transform_target = True,
                   combine_rare_levels = True, rare_level_threshold = 0.05,
                   remove_multicollinearity = True, multicollinearity_threshold = 0.95,
                   bin_numeric_features = ['odometer'],
                   log_experiment = True, experiment_name = 'carsV2')
```

## *CONSTRUCTION ET COMPARAISON DES MODELES*

Je déclare une variable "top3" qui contient les 3 premiers modèles au classement en vue de les combiner plus tard pour en faire un modèle unique

```python
top3 = compare_models(exclude = ['ransac', 'tr'], fold=5, n_select = 3)
```

```python
# Visualisation de la variable ( les modeles SVR, Catboost et Light GBM Regressor sont retenus)
print(top3)
```

***Observations importantes***
- SVR en 1ère position après feature engineering (normalization, multicolinearite, transformation...) avec r2: 0.8208 >-0.0483 (17ème position dans 1ère expérience)
- RMSLE et RMSE du SVR les plus basses tte exp confondue 
- Catboost et XGBoost en 2eme et 4eme position avec métriques plus faibles que xp1
